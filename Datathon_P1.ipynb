```python
# Q1

path_orders = '/content/drive/MyDrive/Datathon/Transaction/orders.csv'
df_orders = pd.read_csv(path_orders)
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])

df_unique = df_orders['customer_id'].value_counts()
df_selected = df_orders[df_orders['customer_id'].map(df_unique) > 1]

unique = df_selected['customer_id'].unique().tolist()
l = []

for val in unique:
  l_temp = []
  df_temp = df_selected[df_selected['customer_id'] == val]
  for i in range(len(df_temp) - 1):
    l_temp.append(np.abs(df_temp.iloc[i]['order_date'] - df_temp.iloc[i+1]['order_date']))
  l.extend(l_temp)

median = np.median(np.array(l))
print(median)


# df_orders.head()
```

```python
#Q2
import pandas as pd

path_products = '/content/drive/MyDrive/Datathon/Master/products.csv'

df_products = pd.read_csv(path_products)

df_products['margin'] = (df_products['price'] - df_products['cogs']) / df_products['price']

ket_qua = df_products.groupby('segment')['margin'].mean().sort_values(ascending=False)

print("Kết quả tính toán cho Câu Q2:")
print(ket_qua)
```

```python
#Q7
import pandas as pd

path_sales = '/content/drive/MyDrive/Datathon/Analytical/sales.csv'
path_geo = '/content/drive/MyDrive/Datathon/Master/geography.csv'
path_orders = '/content/drive/MyDrive/Datathon/Transaction/orders.csv'

df_sales = pd.read_csv(path_sales)
df_geo = pd.read_csv(path_geo)
df_orders = pd.read_csv(path_orders)

sales_with_zip = pd.merge(df_orders, df_sales, left_on='order_date', right_on='Date')
final_data = pd.merge(sales_with_zip, df_geo, on='zip')

ket_qua_q7 = final_data.groupby('region')['Revenue'].sum().sort_values(ascending=False)

print("Tổng doanh thu theo từng vùng (Region):")
print(ket_qua_q7)
```
